```python
#TELANAGANA ROUTES AND LINKS
     
#open brower
driver_t = webdriver.Chrome()

#load webpage
driver_t.get("https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile")

time.sleep(5)

#maximizing window
driver_t.maximize_window()
```

```python
#retrieve bus links and routes
wait = WebDriverWait(driver_t, 30)
def tsrtc_link_routes(path):
    link_tsrtc = []
    route_tsrtc = []
    for i in range(1,4):
        paths = driver_t.find_elements(By.XPATH,path)
        #retrieve route links
        for link in paths:
            L=link.get_attribute("href")
            link_tsrtc.append(L)
        #retrieve route names
        for route in paths:
            route_tsrtc.append(route.text)
        try:
            #scroll the whole page to make sure the pagination container is visible
            #driver_t.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            #time.sleep(3)#let the scrolling take time
            
            #wait for pagination element to be present 
            pagination=wait.until(EC.presence_of_element_located((By.XPATH,"//div[@class='DC_117_paginationTable']")))

            #check for the next button exist
            next_button_xpath = f"//div[@class='DC_117_pageTabs ' and text()='{i+1}']"
            next_buttons = driver_t.find_elements(By.XPATH, next_button_xpath)
            if not next_buttons:
                print(f"No more pages to paginate after page {i}")
                break
            
            #if next button exist make it visible and clickable
            next_button=wait.until(EC.element_to_be_clickable((By.XPATH, next_button_xpath)))
          
            #scroll the next button into view,using javascript
            driver_t.execute_script("arguments[0].scrollIntoView();",next_button)
            
            #click the next button,using javascript
            time.sleep(10)
            driver_t.execute_script("arguments[0].click();",next_button)
            
        except NoSuchElementException as e:
            print(f"No such element at page {i}: {str(e)}")
            break
        except TimeoutException as e:
            print(f"Timed out waiting for elements at page {i}: {str(e)}")
            break
        except Exception as e:
            print(f"An unexcepted error occured at page {i}: {str(e)}")
            break
    return link_tsrtc, route_tsrtc
link_tsrtc, route_tsrtc = tsrtc_link_routes("//a[@class='route']")
```

```python
#Bihar state road transport corporation (BSRTC)
#open browser
driver_bsrtc = webdriver.Chrome()
#load webpage
driver_bsrtc.get("https://www.redbus.in/online-booking/bihar-state-road-transport-corporation-bsrtc/?utm_source=rtchometile")
#maximize window
time.sleep(5)
driver_bsrtc.maximize_window()
```

```python
#retriving links and routes
wait = WebDriverWait(driver_bsrtc, 30)
#writing a funcion 
def bsrtc_links_routes(path):
    links_bsrtc=[]
    routes_bsrtc=[]
    for i in range(1,5):
        paths=driver_bsrtc.find_elements(By.XPATH,path)
        #retrieve links
        for link in paths:
            bsrtc=link.get_attribute("href")
            links_bsrtc.append(bsrtc)
        #retrieve routes
        for route in paths:
            routes_bsrtc.append(route.text)
        try:
            #wait till paginate element is located
            pagination=wait.until(EC.presence_of_element_located((By.XPATH, "//div[@class='DC_117_paginationTable']")))
            #check if next page button exists
            next_page_xpath = f"//div[@class='DC_117_pageTabs ' and text()='{i+1}']"
            next_page_buttons = driver_bsrtc.find_elements(By.XPATH, next_page_xpath)
            if not next_page_buttons:
                print(f"No more pages to paginate after page {i}")
                break
            #if next page button exist make it clickable
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, next_page_xpath)))
            #scroll to the next page button to be visible using javascript
            driver_bsrtc.execute_script("arguments[0].scrollIntoView();", next_button)
            #click the next page button using javascript
            time.sleep(10)
            driver_bsrtc.execute_script("arguments[0].click();", next_button)
        except NoSuchElementException as e:
            print(f"No such element is present at page {i}: {str(e)}")
            break
        except TimeoutException as e:
            print(f"Timed out waiting for element at page {i}: {str(e)}")
            break
        except Exception as e:
            print(f"An unexcepted error occured at page {i}: {str(e)}")
            break
    return links_bsrtc, routes_bsrtc
links_bsrtc, routes_bsrtc = bsrtc_links_routes("//a[@class='route']")
```

```python
df_bsrtc = pd.DataFrame({"bus route":routes_bsrtc,"bus link":links_bsrtc})
df_bsrtc
```
